```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df_train = pd.read_csv('D:/prac/mobile price/train.csv')
df_test = pd.read_csv('D:/prac/mobile price/test.csv')
```

```python
X = df_train.drop('price_range',axis=1)
y = df_train['price_range']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

lr = LinearRegression(n_jobs=-1, normalize=True, fit_intercept=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=111)
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=10)
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtcl = DecisionTreeRegressor()
dtcl.fit(X_train,y_train)
```

```python
dtcl.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=500)
rf.fit(X_train, y_train)
```

```python
rf.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score, KFold
cv = KFold(5, shuffle=True, random_state=11)
cvs = cross_val_score(knn, X_test, y_test, scoring="r2", cv=cv)
cvs
```
